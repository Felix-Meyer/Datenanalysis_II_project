```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
narrow_beam = np.loadtxt('data/optimal_pos.txt')
divergent_kaons = np.loadtxt('data/optimal_pos_divergent_kaons.txt')

plt.figure()
plt.hist(narrow_beam,bins=20)
plt.title(f'Detector position for infinitely narrow beam, N=50, $\sigma=${np.round(np.std(narrow_beam,),1)}')
plt.xlabel('Detector position [m]')
plt.ylabel('Detections')
plt.savefig('plots/optimal_pos_hist_narrow_beam.pdf')
plt.show()

plt.figure()
plt.hist(divergent_kaons,bins=20)
plt.title(f'Detector position for divergent beam, N=50, $\sigma=${np.round(np.std(divergent_kaons,),1)}')
plt.xlabel('Detector position [m]')
plt.ylabel('Detections')
plt.savefig('plots/optimal_pos_hist_divergent_beam.pdf')
plt.show()
```
